Let us first examine the unique values in each column to see which columns have missing values.

````python
def show_num_nulls(df):
    """
    function to show the number of missing entries in each column given a dataframe df
    """
    for c in df.columns:
        num_null = df[c].isnull().sum()
        if (num_null > 0):
            print('Number of null values in {}: {}'.format(c,num_null))

show_num_nulls(fifa19_players_skills)
````

Two aggregation schemes are usually used. We can either fill the missing values with the median of the data or the mean of the data. The distribution we have is left skewed. Hence, it seems like a good idea to use the median of the data to fill the missing values as it is more robust to outliers.

## A3. Building a regression model

By letting them perform exercises, you have gathered data about the following athletic skills (except for the *overall football performance*) of your candidate players:

````python
athletic_skills = ['Overall', 'Acceleration',
                   'SprintSpeed', 'Agility', 'Reactions', 'Balance',
                   'Jumping', 'Stamina', 'Strength']
````

To decide which player to hire, you need to know how these skills transfer to football performance.

To predict the football performance of a player, you want to build a linear model with an additional regularization term to prevent overfitting. The specific model used in this exercise is L2-regularized linear regression, aka "Ridge Regression". It combines the least squares linear regression model that you know from the lecture with a penalty on the norm of the weight vector [1]. The function that Ridge Regression minimizes is

$$||y-Xw||_2^2 + \alpha ||w||_2^2,$$

where $y$ is a vector of overall ratings, $X$ is the matrix of data vectors with each *row* being the features of a football player, $w$ is the weight vector, and $\alpha$ is the (non-negative) regularization parameter.

[1] If there are enough model parameters, a powerful machine learning model will be able to "memorize" the value it is supposed to predict for some training examples. That means, it will perform well on those training examples, but will perform much worse on data that was not used for training. One way to counter this is to penalize large weight values. This can be thought of as reducing the complexity of the model.

### A3.1.
Split the data randomly into a train (70%) and a test set (30%). Use 1 as the random seed.

````python
fifa19_players_athletic_skills = fifa19_players_skills[athletic_skills].copy()
````

````python
#y is a vector of "Overall" ratings
y = fifa19_players_athletic_skills['Overall'].to_numpy()

#X is the matrix of data vectors with each row being the features 
#(from "Acceleration" to "Strentgh") of a football player
X = fifa19_players_athletic_skills.iloc[:,1:].to_numpy()
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=SEED)
````

### A3.2.
Build a Ridge Regression model that predicts a player's overall football performance from the other feature in the list above (sklearn might be helpful here).

Ridge Regression has a regularization parameter $\alpha$. Optimize it in the range [0, 2000] via cross validation. Use as a metric the mean squared error. Plot the cross-validation error as a function of the regularization parameter and choose the best parameter for your final model.

**Solution:**

We stick to using k-fold cross validation to determine the best hyperparameter.

````python
#Setting the parameters
alphas = list(np.arange(0,2001, 1)) #Gives us [0,2000] inclusive in borders
k_folds = [5, 9] # We tried different k values and we chose to show here 5-and 9-fold mainly because of they gave the best results.
````

````python
# Keep cross validation errors by iterating over the alpha and k (for k-fold) hyperparameters
cv_score = []
for cv_k in k_folds:
    for a in alphas:
        clf = Ridge(alpha = a)
        cv_score.append(-cross_val_score(estimator = clf, 
                                         X = X_train, 
                                         y = y_train, 
                                         cv = cv_k,
                                         scoring = "neg_mean_squared_error").mean())
````

````python
cv_score = np.reshape(cv_score, (2,2001)) #Reshape to separate 5-fold and 9-fold results
score = pd.DataFrame(cv_score).T
score.columns = ['cv_score_5k_fold', 'cv_score_9k_fold']
score['alpha'] = alphas
score.set_index(['alpha'])

score
````

````python
fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111)    # The big subplot
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

# Turn off axis lines and ticks of the big subplot
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('none')
ax.spines['left'].set_color('none')
ax.spines['right'].set_color('none')
ax.tick_params(labelcolor='w', top='off', bottom='off', left='off', right='off')

#Set x and y values to plot 
ax1.plot(score.alpha, score.cv_score_5k_fold)
ax2.plot(score.alpha, score.cv_score_9k_fold)

# Set common labels
fig.text(0.5, 0.04, 'Alpha Values', ha='center', va='center')
fig.text(0.06, 0.5, 'MSE', ha='center', va='center', rotation='vertical')

ax1.set_title('5-fold Cross Validation Score')
ax2.set_title('9-fold Cross Validation Score')
````

From the above plot, we can see that for the first plot, the best alpha value should lie in the interval [750,1250], the tuning of the hyperparameter gives us no useful result since the error for larger alpha values are the same. But the plot is not similar to what we saw in the lecture for the 5-fold. On the other hand, for the second plot: the best alpha value seems to lie in the interval [0,250]. Increasing the value of the alpha hyperparameter does not give us a smaller error since the error increases with alpha beyond that interval. Hence, this plot is quite similiar what we saw in the lectures, so we can say that it gives us a reasonable MSE result too.

Let's now examine the best alpha value for the 9-fold cross validation.

````python
#Initialize the parameters
best_alpha = 0
cv_score_0 = score.cv_score_9k_fold[0]

for i in range(1,len(score.cv_score_9k_fold)):
    #We round the score for the last 6 digits, like a small epsilon value to compare the scores
    while (np.round(score.cv_score_9k_fold[i],6)<np.round(score.cv_score_9k_fold[i-1],6)):
        i+=1
        if (score.cv_score_9k_fold[i]==score.cv_score_9k_fold[i-1]):
            break
        best_alpha = score.alpha[i-1]
        break

print('The best alpha value for a 9-fold cross validation on 70% train data is: {}'.format(best_alpha))
````

Let's see the visualisation of our predicted and original football player performance for the best alpha value (117 obtained by 9-fold cross validation).

````python
#Visualize Predicted vs. Original
clf = Ridge(alpha = 117)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0))
ax.plot([np.min(y_train), np.max(y_train)], [np.min(y_train), np.max(y_train)], 'r--', lw=4)
plt.title('Predicted vs. Original')
ax.set_xlabel('Original Overall Football Performance')
ax.set_ylabel('Predicted Overall Football Performance')
plt.show()

cv_score = (-cross_val_score(clf, 
                             X_train, 
                             y_train, 
                             cv=9, 
                             scoring = "neg_mean_squared_error"))

print("As seen from the 'Predicted vs. Original' plot, our Ridge Regression model predicts with MSE score of: {}.".format(np.round(cv_score.mean(),8)))
````

Now we proceed to retrain the models 1000 times. At each iteration we sample from a fixed training set, train the model using the sample and then we compute the MSE on fixed test datasets and record it.

````python
# Function to compute MSE for each iteration of the retraining
def retrain_mse_with_bootstrap(alpha, X_train, y_train, X_test, y_test, random_state):
    """
    function that takes a random sample from X and y, uses it to train a ridge regression model with the given alpha and returns the MSE 
    Params:
    alpha - the hyperparameter to use in ridge regression
    X_train - the inputs and features to use in taining
    y_train - the labels to use in training
    X_test - the inputs and features to test our model on (for mse computation)
    y_test - the labels to use when testing our model
    random_state - the seed to be used when doing the sampling of X, y
    Returns:
    mse - the mean squared error of the computed model
    """
    (m, n) = X_train.shape
    sample = resample(X_train, y_train, n_samples=int(0.7*m), random_state=random_state)
    r = Ridge(alpha=alpha)
    r.fit(sample[0],sample[1])
    y_pred = r.predict(X_test)
    return mean_squared_error(y_test, y_pred)
    
````

````python
# Retrain models using obtained alpha
alpha_model_1 = 0.0826 #optimal alpha from A4.1
alpha_model_2 = 0.6 #optimal alpha from A5.1
model_1_scores = np.zeros((n_iters,)) #to store mse's for model 1
model_2_scores = np.zeros((n_iters,)) #to store mse's for model 2
for i in range(1, n_iters + 1):
    model_1_scores[i-1] = retrain_mse_with_bootstrap(alpha_model_1, X_train_std, y_train, X_test_std, y_test, random_state=i)
    model_2_scores[i-1] = retrain_mse_with_bootstrap(alpha_model_2, X_all_skills_train_std, y_all_skills_train, X_all_skills_test_std, y_all_skills_test, random_state=i)
````

Having said that, the goal is to discover similar subreddits by only looking at the *words* present in the posted messages. Once such a list of similar subreddits is identified, an appropriately chosen moderator can then be assigned to all these subreddits.

Specifically, the task boils down to computing a similarity score between two subreddits based on the *words* present in their textual content. Your first idea is to use the *Jaccard similarity*, which is defined as the size of the intersection of two sets divided by the size of their union.

$Jaccard(A,B) = \frac{|A \cap B|}{|A \cup B|}$

````python
def jaccard_similarity(list1, list2):
    s1 = set(list1)
    s2 = set(list2)
    return len(s1.intersection(s2)) / len(s1.union(s2))
````

### B2.1.
The first step requires constructing a set representation of each subreddit. The goal is to represent each subreddit as a *set of words* existing in the messages posted on that subreddit. Compute the 50,000 most frequent words across all the provided subreddits. Construct a representation for each subreddit by retaining only the words found in the previously identified set of 50,000 frequent words.

Some rules:
 * Words are defined as tokens matching the regular expression `\W`
 * Remove all the stop-words (English language)

*Note: You might find the [RegexTokenizer](https://spark.apache.org/docs/2.2.0/ml-features.html#tokenizer) and the [StopWordsRemover](https://spark.apache.org/docs/2.2.0/ml-features.html#stopwordsremover) utilities available in the package pyspark.ml useful for this task as they help you in transforming the features and removing stopwords.*

As indicated in the note, we use the RegexTokenizer and the StopWordsRemover available in the package pyspark.ml

````python
# Here, we only use reddit_posts 
# Let's compute the number of usages for each word in the posts
posts_subreddit = reddit_posts.drop('author', 'id')

regexTokenizer = RegexTokenizer(inputCol="body", outputCol="words", pattern="\\W")
regexTokenized = regexTokenizer.transform(posts_subreddit)

#remove the stopwords
remover = StopWordsRemover(inputCol="words", outputCol="filtered_words")
filtered_posts = remover.transform(regexTokenized)
filtered_posts = filtered_posts.drop('body', 'words')
# each row in filtered_posts is a message posted in column 'subreddit'.
# The column 'filtered_words' contains the relevant words for that particular message

# each row in the df all_words is a word and the subreddit where it was used
all_words = filtered_posts.select('subreddit', explode('filtered_words').alias("word"))
all_words.show(10)
````

````python
# We count how many times each word was used, sort in descending order,
# drop the count column and take the 50'000 first words
frequent_words = all_words.groupBy("word").agg(count("*").alias("w_count")).sort(desc("w_count")).limit(50000)
````

````python
# Construct representation for each subreddit by retaining only the words found in the most frequent 50,000 words

# We can filter out the words that do not belong to the most frequent 
# 50'000 words from the dataframe all_words (exploded)
all_words_filtered = all_words.join(frequent_words, 'word', 'left_semi')
all_words_filtered.show(10)
````

````python
# We use collect_set, which is the "inverse" of explode
words_sub = all_words_filtered.groupBy('subreddit').agg(collect_set('word').alias('words'))
print('Representation for each subreddit by retaining only the words found in the set of 50\'000 frequent words : ')
words_sub.show(10)
````

However, the Jaccard measure is inversely proportional to the union of both sets which will be a big number. The final result will thus be a small number, even if both sets were very wimilar.
